```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis", model="finiteautomata/bertweet-base-sentiment-analysis")
```

```python
classifier("If you leave me, what can I do?")
```

```python
classifier("I'm sad that I couldn't say hello more brightly")
```

```python
classifier("Thank you. Be happier than I am.")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load the tokenizer and model from Hugging Face
model_name = "sbcBI/sentiment_analysis_model"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Fungsi untuk melakukan prediksi sentimen
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model(**inputs)
    predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
    labels = ['negative', 'neutral', 'positive']
    predicted_label = labels[torch.argmax(predictions)]
    return predicted_label, predictions
```

```python
sentence = "If you leave me, what can I do?"
label, scores = predict_sentiment(sentence)

print(f"Sentence: {sentence}")
print(f"Predicted Sentiment: {label}")
print(f"Confidence Scores: {scores}")
```

```python
sentence = "I'm sad that I couldn't say hello more brightly"
label, scores = predict_sentiment(sentence)

print(f"Sentence: {sentence}")
print(f"Predicted Sentiment: {label}")
print(f"Confidence Scores: {scores}")
```

```python
sentence = "Thank you. Be happier than I am."
label, scores = predict_sentiment(sentence)

print(f"Sentence: {sentence}")
print(f"Predicted Sentiment: {label}")
print(f"Confidence Scores: {scores}")
```

Hasil perbandingan prediksi sentimen tanpa (model awal) dan dengan Attention Transformet menggunakan model Pre-Trained BERT:

1. **"If you leave me, what can I do?"** --> Hasil prediksi model awal (**Netral, score : 0,499**) hampir seimbang antara sentimen netral dan lainnya, ini menunjukkan bahwa model kurang bisa menangkap emosi yang jelas. Dengan attention, model lebih yakin bahwa sentimen netral lebih dominan (**Netral, score : 0,6617**), mungkin karena fokus pada kata-kata yang mengekspresikan kebingungan tanpa emosi negatif yang kuat.  

2. **"I'm sad that I couldn't say hello more brightly"** --> Hasil prediksi model awal (**Negatif, score : 0,839**) yang dimana model yakin ini negatif karena adanya ekspresi kekecewaan dan kesedihan. Dengan attention, meskipun tetap negatif, score model berkurang (**Negatif, score : 0,5730**) yang mungkin karena model lebih fokus pada frasa "say hello more brightly" yang sedikit melemahkan kesan negatif secara keseluruhan.

3. **"Thank you. Be happier than I am."** --> Hasil prediksi model awal (**Positif, score : 0,992**) yang artinya model sangat yakin bahwa ini kalimat positif dengan adanya frasa "Thank you." Dengan attention, meskipun model tetap yakin bahwa kalimat ini positif (**Positif, score : 0,9489**), ada sedikit penurunan score karena mungkin perhatia nmodel lebih merata terhadap seluruh kalimat, termasuk frasa "than I am."

Kesimpulannya bahwa dengan attention transformer, model menjadi lebih baik dalam menangkap konteks secara menyeluruh dan memberikan prediksi yang lebih seimbang, terutama pada kalimat yang memiliki campuran emosi atau struktur yang lebih kompleks.
